```python
species_column = df_fish['Species'].value_counts()
plt.bar(species_column.index, species_column.values)
plt.xlabel('Species')

plt.ylabel('Values')

plt.title('Bar Chart for Fish Species')
plt.xticks(rotation=45)

plt.show()
```

This code uses Matplotlib to create a bar chart that visualizes the distribution of fish species in a DataFrame called df_fish. It customizes the plot by adjusting figure size, adding colors.On x axis we took species and on y-axis values with rotation as 45.Then we created graph gave a title as Bar Chart for Fish Species

```python
column_to_plot = ['Height' , 'Width']

plt.figure(figsize=(10, 6))

df_fish[column_to_plot].boxplot()

plt.title('Box Plots for Height & Width Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)

plt.show()
```
